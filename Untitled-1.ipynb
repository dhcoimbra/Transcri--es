```python
from transformers import DonutProcessor, VisionEncoderDecoderModel
from PIL import Image
import torch
from transformers import LayoutLMv2Processor, LayoutLMv2ForTokenClassification
```

```python
# Carregar o processador e o modelo
YOUR_TOKEN = "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"

#processor = DonutProcessor.from_pretrained("kakaobrain/donut-base", use_auth_token=YOUR_TOKEN)
#model = VisionEncoderDecoderModel.from_pretrained("kakaobrain/donut-base", use_auth_token=YOUR_TOKEN)

processor = LayoutLMv2Processor.from_pretrained("microsoft/layoutlmv2-base-uncased")
model = LayoutLMv2ForTokenClassification.from_pretrained("microsoft/layoutlmv2-base-uncased")



def load_image(image_path):
    # Carregar a imagem
    return Image.open(image_path).convert("RGB")

def identify_receipt(image_path):
    # Carregar e processar a imagem
    image = load_image(image_path)
    pixel_values = processor(image, return_tensors="pt").pixel_values

    # Fazer a inferência
    output = model.generate(pixel_values)
    decoded_output = processor.batch_decode(output, skip_special_tokens=True)[0]

    # Verificar se há palavras-chave relacionadas a um comprovante
    if any(keyword in decoded_output.lower() for keyword in ["comprovante", "pagamento", "depósito", "transferência"]):
        print(f"Comprovante bancário identificado na imagem: {image_path}")
    else:
        print(f"Nenhum comprovante identificado na imagem: {image_path}")

# Exemplo de uso
```
